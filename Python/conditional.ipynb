**Conditional Statements :**
- conditional are used to execute the block of code when you meet with certain conditions
- it iterates only once and used to make decisions on certain conditions
- there are following condiitonal statements in python :
     1. if
     2. if-else
     3. if-elif-else
     4. nested if

**if :** 
- used to execute the code only when a condition is true
- if conditional is false then happens nothing
- used one there is only one possible way

```python
lst = [1,2,3,4,5]
if 2 in lst:
    print("yes")
```

**if-else :** 
- if - executes when condition is true
- else - executes when if condition is false
- used one there is two possible way

```python
lst = [1,2,3,4]
if 2 in lst: # true
    print("yes")
else:
    print("no")    
```

```python
lst = [1,2,3,4]
if 5 in lst: # false
    print("yes")
else:
    print("no") 
```

**if-elif-else :**
- used when you have multiple conditions
- if -- execute when condtion is true
- elif --  execute when if condition is false
- else -- execute when both if and elif condition is false

```python
lst = [1,2,3,4,5,6,7]
lst1 = [1,2,3,4,5,6,7]
if 2 in lst: # true
    print("2 is in lst")
elif 2 in lst1:
    print("2 is in lst1") 
else:
    print("2 is not either in lst or lst1")       
```

```python
lst = [1,3,4,5,6,7]
lst1 = [1,2,3,4,5,6,7]
if 2 in lst: # false
    print("2 is in lst")
elif 2 in lst1: # true
    print("2 is in lst1") 
else:
    print("2 is not either in lst or lst1")   
```

```python
lst = [1,3,4,5,6,7]
lst1 = [1,3,4,5,6,7]
if 2 in lst: # false
    print("2 is in lst")
elif 2 in lst1: # false
    print("2 is in lst1") 
else: #
    print("2 is not either in lst or lst1")   
```

**Nested-if/if-else/if-elif-else:**
- used when you have condition in condition

```python
# nested-if
lst = [1,2,3,4,5,6,7]
lst1 = [1,2,3,4,5,6,7]
if 2 in lst:
    if 2 in lst1:
        print("2 is in both lst and lst1")
```

```python
# nested-if-else
lst = [1,2,3,4,5,6,7]
lst1 = [1,3,4,5,6,7]
if 2 in lst:
    if 2 in lst1:
        print("2 is in both lst and lst1")
    else:
        print("2 is in lst not in lst1")    
else:
    if 2 in lst1:
        print("2 is in lst1 not in lst")
    else:
        print("2 neither in lst nor in lst1")       
```

```python
# nested-if-elif-else
lst = [1,3,4,5,6,7]
lst1 = [1,2,3,4,5,6,7]
if 2 in lst:
    if 2 in lst1:
        print("2 is in both lst and lst1")
    else:
        print("2 is in lst not in lst1")   
elif 2 in lst1:
    if 2 in lst:
        print("2 is in both lst and lst1")
    else:
        print("2 is in lst1 not in lst") 
else:
    print("2 neither in lst nor in lst1")                  
```
