```python
## Question_1 : Dictionary 
a=int(input("define no of items in dictionary"))
nk={}
for i in range(a+1):
    key=input("enter key: ")
    if(key in nk):
        print("key already exists")
        continue
    value=input("enter value: ")
    nk[key]=value
print(nk)  
```
