```python
df = sns.load_dataset('tips') ## tips is a seaborns defualt data sets
```

```python
sns.lineplot(df['total_bill'])
plt.show()
```

- catplot
- relplot : define relation

```python
sns.relplot(x=df['total_bill'],y = df['tip'])
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=df)
plt.show()
```

```python
## time,total_bill,
sns.relplot(x="total_bill",y="tip",hue='time',style='smoker',data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style='smoker',kind='scatter',row='time',data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style='smoker',kind='scatter',col='time',data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style='smoker',kind='scatter',row='smoker',data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style='smoker',kind='scatter',col='day',data=df)
plt.show()
```

```python
df['day'].unique()
```

```python
sns.relplot(x="total_bill",y="tip",data=df)
plt.ylabel('tips',color='red',fontsize=20)
plt.xlabel("total_bill" ,color='red',fontsize=20)
plt.title("graph btw tips and total_bill",color='red',fontsize=20)
plt.legend()
plt.show()
```

catplot

```python
sns.catplot(y = 'total_bill',x='day',kind='strip',data=df)
plt.show()
```

```python
sns.catplot(y = 'total_bill',x='day',kind='boxen',data=df)
plt.show()
```

```python
sns.catplot(y = 'total_bill',x='day',kind='bar',data=df)
plt.show()
```

```python
sns.catplot(y = 'total_bill',x='day',kind='violin',data=df)
plt.show()
```

```python
sns.catplot(y = 'total_bill',x='day',kind='swarm',data=df)
plt.show()
```

```python
sns.catplot(y = 'total_bill',x='day',kind='box',data=df)
plt.show()
```
