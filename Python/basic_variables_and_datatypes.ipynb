Declaration of Variables : 

variables : containers to store data value (value can be integer, string,float,complex,boolean etc.)
variables can be defined using alphabets(a-z,A-Z),numbers(0-9)and underscore(_ or __)

Rules to declare variables :
  1. starts with either alphabets(a-z,A-Z) or underscores(_ or __).
  2. don't use numbers at starting.
  3. don't use reserved words as variables.
  4. don't use space in between variable name.
  

```python
# variable containing alphabet only
a=10   
print(a)
# variable containing alphabet and number
a1=50
a1a=20
# 1a = 40   ## wrong way to declare variable
print(a1)
print(a1a)
#variable containing alphabet and underscores
_a=90  # starts with single underscore
__a=80 # starts with double underscore
a_b=70 #contains underscore in between
a_=60 # ends with underscore
print(_a)
print(__a)
print(a_b)
print(a_)

```

** Question: why we use single quotes('') and double quotes("") to define single line string?

7. List : collection of heterogeneous data element separated by "," and enclosed within square brackets "[]" 
- mutable : we can manipulate or modify the liet elements
- ordered : stores the element in the same order as it is inserted/append in the list
- indexex : access list element using index
- allow duplicate elements

```python
lst = [1,2,'abc',3.5,True]
print(type(lst))
print(lst)
print(lst[0]) ## access element of the list using index number
```

List Operations :


```python
lst=[1,'avc',3.5,True,1]
print(lst)
print(len(lst)) ## return the number of elements in the list
lst.append(2+3j)  ## used to add element at the end of the list
print(lst)
lst.append([1,2,3]) ## used to add list as an element at the end of the list
print(lst)
lst.insert(2,'python') ## used to add element at specific index in the list list.insert(index,element)
print(lst)
lst.pop() ## used to remove last element from the list
print(lst)
lst.pop(2) ## used to remove element from specific index in the list list.pop(index)
print(lst)
lst.remove('avc') ## used to remove specific element from the list list.remove(element)
print(lst)
lst.reverse() ## used to reverse the sequence of elements of the list
print(lst)
print(lst.count(1)) ## return the no of specified element in the list
print(lst.index(3.5)) ## return the index of specified element in the list
print(lst.index(1)) ## return the index of first occurrence of specified element in the list
```

```python
lst=[1,2,3,4,5]
lst1=lst.copy() ## used to create a copy of the list
print(lst1)
lst2=lst ## used to create a reference of the list 
print(lst2)
lst3=lst.deepcopy(x) ## used to create a deep copy of the list
print(lst3)
lst.pop()
print(lst1)
print(lst2)
print(lst3)
```

list operations on numeric list :


```python
lst1=[1,2,3,4,5.6,3,0]
print(sum(lst1)) ## return the sum of list elements
print(max(lst1)) ## return the max element from the list
print(min(lst1)) ## return the min element from the list
lst1.sort() ## used to sort the list in ascending order
print(lst1)
## another way to sort
lst1.sort(reverse=True) ## used to sort the list in descending order
print(lst1)
print(sorted(lst1)) ## return a new sorted list without modifying the original list
print(sorted(lst1,reverse=True)) ## return a new sorted list in descending order without modifying the original list
```

packing and unpacking of tuple elements

```python
a=4
b=6
c=8
tup=(a,b,c) # packing of values in tuple
print(tup)
c,d,e=tup # unpacking of values from tuple
print(c)
print(d)
print(e)
```

Tuple operations :

```python
tupl=(1,2,3,1)
print(tupl)
print(len(tupl)) ## return the number of elements in the tuple
print(tupl[2]) ## access element of the tuple using index number
print(tupl.count(1)) ## return the no of specified element in the tuple
print(tupl.index(3)) ## return the index of specified element in the tuple
print(tupl.index(1)) ## return the index of first occurrence of specified element in the tuple  
print(sum(tupl)) ## return the sum of tuple elements
print(max(tupl)) ## return the max element from the tuple
print(min(tupl)) ## return the min element from the tuple   
tupl2=tupl ## we can't copy tuple but give reference of the tuple to another tuple variable.
print(tupl2)
```
