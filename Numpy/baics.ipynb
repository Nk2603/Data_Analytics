```python
## converts list --> array
lst = [1,2,3,4]
print("my list : ",lst)
print("Type of my list : ",type(lst))
arr = np.array(lst)
print("my array : ",arr)
print("Type of my array : ",type(arr))
```

```python
## list using range function
lst = list(range(8)) # define a range and then typecast it to list
# range() used to get element between specific range
print(lst,type(lst))
arr = np.array(lst)
print(arr,type(arr))
```

```python
arr = np.arange(2,22) ## array using range function 
print(arr,type(arr))
```

```python
arr = np.zeros(10) ## array of 0 and 10 is no of elements 
print(arr)
# 2-D np.zeros((rows,columns))
arr1 = np.zeros((5,10)) ## 5 is number of rows and 10 is number of columns
print(arr1)
print(arr1)

print(type(arr[1])) ## type of element is float that's why is goves output as 0.
```

```python
arr = np.ones(10) ## array of 1
print(arr)
# 2-D
arr1 = np.ones((5,10)) ## 5 is number of rows and 10 is number of columns
print(arr1)
```

```python
arr2 = np.linspace(1,10,5) ## 1 to 10 is a range and total elements is 5
print(arr2)
```

converts 1-D array to 2-D array using reshape()

```python
arr = np.arange(0,30)
print(arr)
# 1-D to 2-D reshape is used to change the dimensions of the arrayd
arr = arr.reshape(5,6) ## 5 is rows and 6 is columns
print(arr)
```

```python
# random : return random numbers within defined range
# random.randint() : random numbers are of integer type
arr = np.random.randint(1,5,10) ## range = 1 to 5 and 10 is total elements
arr
```

```python
arr = np.random.randint(1,5,(2,5)) ## range = 1 to 5 and 2 is row and 5 is column and 2*5=10 are elements
arr
```

```python
# random.random() :  random numbers are of float type
# - default range is 0.0 to 1.0
arr = np.random.random(10) ## 1-D
arr
```

```python
arr = np.random.random((3,5)) ## 2-D 
arr
```

3-D Array

```python
arr3 = np.arange(225).reshape(5,15,3) ## (tables,rows,columns) 
  ## 5 -> tables , 15 -> rows , 3 -> columns
arr3
```

```python
print("shape : ",arr3.shape)
print("data type : ",arr3.dtype)
print("size : ",arr3.size)
print("dimensions : ",arr3.ndim)
```

indexing and slicing in 3-D

```python
print(arr3[0])
```

```python
print(arr3[0][0])
```

```python
## first element of each row of each table
print(arr3[0:5,0:15,0])
```

```python
## first element of each table
print(arr3[0:4,0,0])
print(arr3[0:4][0][0]) # doesn't work properly why ?
```
