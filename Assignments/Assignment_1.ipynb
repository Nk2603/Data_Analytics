Ques-1 : Why are dictionary keys required to be immutable in Python? Explain with a small example


Answer : Dictionary Keys are immutable because dcitionary uses hashing (hash based data structures). 
- Hashing is the process to generate a fixed hash value (numeric-identity) for key to locate/track quickly in the memory.
- If key changes then hash value for that key also changes and dictionary loses its ability to locate/track that key in the memory

example :

```python
## keys are mutable type : list,set,dictionary
# key={1:4,2:6}
Key=[1,2,3]
Dict = {}
Dict[Key]="hello" # --> gives TypeError of unhashable type : 'list' because keys are mutable type
print(Dict)
```

```python
## keys are immutable type : int,string,float,tuple
Key=(1,2,3)
Dict = {}
Dict[Key]="hello"
print(Dict)
print(hash(Key))  ## hash() used to find the hash value of the key
```

Ques-2 : Write a program that takes a list of elements and creates a dictionary showing frequency of each element

```python
lst = [1,3,4,1,2,8,4,5,3,9,0,2,5,1]
print("list : ",lst)
Freq_dict={}
for i in lst:
    if i in Freq_dict:
        continue # continue is keyword used to skip current iteration and moves to next iteration
    else:
        Freq_dict[i]=lst.count(i) 
print("dictionary showing list element frequency : ",Freq_dict)
```

Ques-3 : What is the difference between dict.get(key) and dict[key]?  Write a code example where get() is safer

Answer : Both are used to get the value of key but 
1. dict[key] : 
    - dict[key] raises KeyError if key is not exists in the dictionary
    - better use if you know key exists or not
2. dict.get(key) :
    - dict.get(key) returns none value or specified default value (dict.get(key,default_value))
    - safer to use when you are not sure about key existence    

```python
dict={'Name':'NK','Age':20,'Degree':"B.Tech"}
print(dict)

# dict[key]
print(dict['Degree']) # --> output : B.Tech
print(dict['Branch']) # --> KeyError as 'Branch' is not exits in dictionary 'dict'
```

```python
# dict.get(key,default)
print(dict.get('Degree')) # --> output : B.Tech
print(dict.get('Branch')) # return none as key not exist
print(dict.get('Branch','Not Specified')) # return specified default value 
```

Ques-4 : Write a program to merge two dictionaries and resolve common keys by keeping the large value

```python
dict1 = {'a':65,'b':35,'c':48,'d':78,'e':34}
dict2= {'a':63,'b':39,'c':50,'d':76,'f':40}
max_dict={}
for key in dict1 | dict2: # key(iterator) is the key which is exist either in dict1 or dict2
    if key in dict1 and key in dict2:  # if key exist in both dict1 and dict2
        max_dict[key] = max(dict1[key],dict2[key])  # max() return the max value
    elif key in dict1:  # key  exists in dict1
        max_dict[key] = dict1[key]
    else:       # key exists in dict2   
        max_dict[key] = dict2[key]
          
             
print(max_dict)   
```
