**LOOPING STATEMENTS** : 
- control statements that allow to run a block of code to run repeatedly as long as condition satisfied 
- it reduce the code of lines if you want to do same task multiple times 
- In Python, there are two types of control/loop statements :
    1. for loop : used when you know exact number of repeatition 
    2. while loop : used when you are not sure about number of repeatition just stopping condition 

*1. For loop :*
- syntax :

      for i in sequence:

          //code
- here i is a iterator/variable which iterates throughout loop 
- sequence can be list,tuple,dictionary,string,set,range        

```python
# for loop for sequence list
lst=[1,2,3,4,5]
for i in lst:
    print(i) # print each element of lst
```

```python
# for loop for sequence tuple
tup=(1,2,3,4)
for i in tup:
    print(i)
```

```python
# for loop for sequence set
sets={1,2,3,4,1}
for i in sets:
    print(i)
```

```python
# for loop for sequence dictionary
dict={'name':"nk",'age':20}
for i in dict:
    print(i) # here i only iterates through list of keys only 
```

```python
dict={'name':"nk",'age':20}
for i in dict.items():
    print(i) # now here i iterate through items(key-value) pair of dictionary
```

```python
dict={'name':"nk",'age':20}
for i in dict.values():
    print(i) # i iterate through list of values  
```

```python
# for loop for sequence string:
string="hello, guys"
for i in string:
    print(i)
```

range() :  range is a build data type or built-in function used to generate/get the sequence of numbers mainly with loops

```python
# for loop for sequence range()
for i in range(5): # 5 is a stopping condition , by default range start from 0 and step is 1
    #i iterates till 5-1, when condition meet i ==5 then loop terminates
    print(i)
```

```python
for i in range(1,5): # starts from 1 and ends at 5-1 = 4 
    print(i)
```

```python
for i in range(1,10,2): # start from 1 , ends at 10-1 = 9 and steps are 2 that means take every 2 number 
    print(i)
```

**list Comprehension**
- we can assign elements to the list using for loop
- short and compact syntax of a for loop

```python
lst=[n for n in range(1,10,2)] # assign odd numbers between 1 and 10 to the list
lst
```

```python
## but there is no such tuple comprehension as it is generator object
tup=(i for i in range(1,10,2))
print(type(tup))
```

```python
# but we can typecast it into tuples:
tup=tuple(tup)
type(tup)
```

Similarly dictionary comprehension and set comprehension can possible as list comprehension

**2. While Loop :**
- syntax :

          while condition : 
              //code
              // update steps like i+=1,i-=1,etc

- executes a block of code as long as condition is true   
- in while loop you must initialize the variable first that comes in use in defining condition for loop            

```python
while i<5:  # when you don't initialize the variable , it doesn't give any error as well as not return any output
    print(i)
    i+=1
```

```python
i=0 # when you  initialize the variable
while i<5:
    print(i)
    i+=1
```

**Other control statements :**
1. break : terminates /exit the loop when iterator with some specified conditions
2. continue : skips the current iteration and moves on next iteration
3. pass : just a placeholder or syntax filler used when you are not sure what to do when you meet with some condition

```python
# break
for i in range(1,10,2):
    if(i==5):
        break # exit the loop when i==5
    print(i) 
```

```python
# continue
for i in range(1,10,2):
    if(i==5):
        continue # skip the current iteration when i==5
    print(i) 
```

```python
# pass
for i in range(1,10,2):
    if(i==5):
        pass # placeholder moves to the next line of block of code not skip the iteration
    print(i) 
```
