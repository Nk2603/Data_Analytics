**EXCEPTION HANDLING :**

- Exception : runtime errors 
- Exception Handling : handle runtime errors so that the program does not crash and your program run safely even if an error occurs

example

```python
a = 10/0
print(a)
```

Here program crash and it raises a runtime error. To continue this program, use exception handling 

***KEYWORDS* In Exception Handling** : 
1. try : block where we write code that may cause error . try block run always but if runtime error occurs it gives that error to except block
2. except : block that handles the try block runtime error
3. else : block that runs if no error occurs or try block runs successfully without any runtime errror
4. finally : block that always runs. This blocks tells that program execution is successfully no matter runtime error comes or not it runs

**try-except**

-> exception handling for non-specified exception or when you are not sure about exception

```python
a=int(input("enter a number: "))
b=int(input("enter a number: "))
try:
    print(a/b)
except Exception as e:
     print(e)
```

-> exception handling for specific exception or when you are sure about exception

```python
try:
    x = int(input())
    y = 10 / x
    print(y)
except ZeroDivisionError:  # when x = 0 
    print("Division by zero")
except ValueError: # when x is anything except number
    print("Invalid input")
```

**try-except-finally**

```python
try:
    x = int(input())
    y = 10 / x
    print(y)
except ZeroDivisionError:  # when x = 0 
    print("Division by zero")
except ValueError: # when x is anything except number
    print("Invalid input")
finally: # run always to mark program execution completion
    print("code execution completed")    
```

**try-except-else-finally**

```python
try:
    x = int(input())
    y = 10 / x
    print(y)
except ZeroDivisionError:  # when x = 0 
    print("Division by zero")
except ValueError: # when x is anything except number
    print("Invalid input")
else: ## when try is success
    print("try block runs successfully")    
finally: # run always to mark program execution completion
    print("code execution completed") 
```
