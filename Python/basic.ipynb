```python
# for = 10  ## cannot use reserved keywords as variable names
# print(for)
# print=10
# print(print)  ## this will not print 10, instead it will print the built-in function 'print' because we have assigned 10 to the variable 'print' which is now overshadowing the built-in function 'print'
del(print)
```

```python
print(10)
a=10 ## start with alphabet, can contain alphabets, digits and underscores, cannot contain spaces, cannot use reserved keywords as variable names
print(a)
_a=8 ## start with underscore
print(_a)
__a=9 ## start with double underscore
print(__a)
a1=7 ## can contain digits
print(a1)
# 1a=7 ## cannot start with digits
a_1=6 ## can contain underscores between digits and alphabets
print(a_1)
```

```python
b=10
b  ## allowed by python but not recommended, it will not print the value of b, instead it will print the built-in function 'b' because we have assigned 10 to the variable 'b' which is now overshadowing the built-in function 'b'
del(b) ## to delete the variable 'b' and restore the built-in function 'b
```

Exceptional Handling
try,except,else,finally

```python
a=int(input("enter a number: "))
b=int(input("enter a number: "))
try:
    print(a/b)
except Exception as e:
     print(e)
finally:
    print("code execution completed") 
```
