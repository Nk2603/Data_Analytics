**FUNCTIONS :**

- Functions are a block of code that performs a specific task and runs only when we call it.
- Functions help to reuse code and avoid repetition.
- A function is created using the def keyword.
- A function can take input (parameters) and can return a result using return.
- syntax :

            def func_name(parameter-1,parameter-2,----,parameter-n):
            
                // code

            func_name(argument-1,argument-2,----,argument-n)           
                        

*Note* : use function name that starts with small letter(snake_case) as class use the first letter capital(Pascalcase) so that classes and functions are differentiated clearly 

**TYPES OF FUNCTIONS :**

1. function with no argument/parameter and no return

```python
def hello():
    print('hello')
hello()    
```

2. functiion with arguments and no return

```python
def sum(a,b): ## here a and b are parameters (placeholders for actual values)
    print("sum : ",a+b)

a=3
b=4
sum(a,b) ## here a and b are arguments (actual values)
# sum(3,4)  ## you can also pass direct values  
```

3. function with arguments and return


```python
def sum(a,b): 
    return a+b

a=3
b=4
print("sum : ",sum(a,b))
```

------> function with multiple return values . In pyhton it's possible that function can return multiple values

```python
def calc(a,b): 
    return a+b,a-b # multiple value

a=3
b=4
summ,diff=calc(a,b)  # a stores in summ and b in diff (store return values in order )
print("sum :",summ)
print("difference : ",diff)
```

4. function with no arguments and return  (rare-case)

```python
def greet():
    return 'Ram Ram'

print(greet())

```

**TYPES OF PARAMETERS :**

1. Positional Parameters  :
- order of arguments == order of parameters

```python
def sum(a,b): ## a for 2 and b for 3 and so on
    return a+b
sum(2,3)   

```

2. Keyword Parameters :
- arguments same keyword as parameters to pass value
- order of arguments and parameter may or may not the same

```python
def sum(a,b): ## a for 5 and b for 2 and so on
    return a+b
sum(b=2,a=5) # b,a --> keywords
```

3. Default Parameters : 
- Default Parameters use a  default value  when no arguments value is given.


```python
def incr(a,b=1):  ## b is a default parameter
    return a+b
incr(2)
```

4. Variable-length Parameters : 
- parameters recieves multiple no. of argument values
- use * preceeding to parameter name

```python
def summ(*num):
    return sum(num)
summ(10,20,30)
```

5. Keyword variable-length parameters :
- parameter recieves multiple keyword argument values
- parameter preceeding by **

```python
def student(**data):
    print(data)

student(name="Asha", age=20, city="Jaipur")
```
