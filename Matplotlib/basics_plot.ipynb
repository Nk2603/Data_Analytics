```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**matplotlib** is powerful library used for data visualization
1. Graph
2. plot
3. scatter
4. hist
5. bar
6. pie

```python
x=np.array([1,2,3,4])
y=np.array([2,4,6,8])
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) ## create a graph gives object
plt.show() ## display the graph
```

```python
plt.plot(x,y,'o') ## o --> point/dot
```

```python
plt.plot(x,y,marker='o') # draw line through dots
```

```python
# x=np.array([1,2,3,4])
y=np.array([2,4,6,8])
plt.plot(y,marker='o')
plt.show()

```

```python
y=np.array([2,4,6,8])
plt.plot(y,marker='*')
plt.show()
```

```python
y=np.array([2,4,6,8])
plt.plot(y,marker='$') ## limited symbols used
plt.show()
```

```python
y=np.array([2,4,6,8])
plt.plot(y,'o:b') ## o --> dot b--> dotted line
plt.show()
```

```python
y=np.array([2,4,6,8])
plt.plot(y,marker='o',ms=20) # incease size
plt.show()
```

```python
y=np.array([2,4,6,8])
plt.plot(y,marker='o',mec='red') #dot outer line colour change mec(marker edge color)
plt.show()
```

```python
plt.plot(y,marker='o',mfc='pink') # marker face color fill the dot 
plt.show 
```

```python
x=np.array([1,2,4,5])
plt.plot(x,ls='dashed')
plt.show()
```

```python
x=np.array([1,2,4,5])
plt.plot(x,ls='dotted')
plt.show()
```

```python
x=np.array([1,2,4,5])
plt.plot(x,linewidth=15) # increase the width of the line
plt.show()
```

```python
x=np.array([1,2,4,5])
plt.plot(x,color='green')
plt.show()
```

```python
# double loine graph
x=np.array([1,2,4,5])
y=np.array([2,3,4.5])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x=np.array([1,2,4,5])
y=np.array([2,3,4,5])
plt.xlabel("x-price")
plt.ylabel("y-label")
plt.title("graph")
plt.plot(x,y)
plt.show()
```
