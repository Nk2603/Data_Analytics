lower v/s casefold :

- lower() : converse normal characters into lower case
- casefold() : converse normal case as well as handle special characters for comparison

```python
print("ß".lower())     # 'ß'
print("ß".casefold())
```

capitalize v/s title :
- capitalize() : converts index 0  to uppercase and all other characters to lower   
- title() : converts first character of each word present in a string to uppercase and others are in lower case.
sometimes behaves a little strangely with apostrophes or special cases

```python
print("hi@s gUy's".capitalize())
print("hi@s gUy's".title())
```

copy v/s deepcopy:
- copy(): built-in method,create new object and shares inner objects
- deepcopy() : not a built-in method , need to import module copy. create new object but doesn't shares inner objects

```python
lst=[1,2,3,4]
print(lst)
lst1=lst.copy()
print("copy :",lst1)
lst2=lst.deepcopy()
print("deepcopy :",lst2)
lst[1]=10
print("copy :",lst1)
print("deepcopy :",lst2)
```

- both gives same result until there is no nested object

```python
import copy
lst=[1,2,3,4]
print(lst)
lst1=copy.copy(lst)
print("copy :",lst1)
lst2=copy.deepcopy(lst)
print("deepcopy :",lst2)
lst[1]=10
print("copy :",lst1)
print("deepcopy :",lst2)
```

difference shown when you used nested objects :
- copy() : create new object but shares inner list 
- deepcopy() : create new objects and copy inner list


```python
import copy
lst=[[1,2],[3,4]]
print(lst)
lst1=copy.copy(lst)
print("copy :",lst1)
lst2=copy.deepcopy(lst)
print("deepcopy :",lst2)
lst[1][0]=10
print("copy :",lst1)
print("deepcopy :",lst2)
# proof that copy() used shared inner list 
print(id(lst[1]), id(lst1[1]), lst[1] is lst1[1])
# proof that deepcopy() doesn't shared inner list
print(id(lst[1]), id(lst2[1]), lst[1] is lst2[1])
```
