Declaration of Variables : 

variables : containers to store data value (value can be integer, string,float,complex,boolean etc.)
variables can be defined using alphabets(a-z,A-Z),numbers(0-9)and underscore(_ or __)

Rules to declare variables :
  1. starts with either alphabets(a-z,A-Z) or underscores(_ or __).
  2. don't use numbers at starting.
  3. don't use reserved words as variables.
  4. don't use space in between variable name.
  

```python
# variable containing alphabet only
a=10   
print(a)
# variable containing alphabet and number
a1=50
a1a=20
# 1a = 40   ## wrong way to declare variable
print(a1)
print(a1a)
#variable containing alphabet and underscores
_a=90  # starts with single underscore
__a=80 # starts with double underscore
a_b=70 #contains underscore in between
a_=60 # ends with underscore
print(_a)
print(__a)
print(a_b)
print(a_)

```

9. Dictionary : collection of heterogeneous items(key:value pair) enclosed within curly braces "{}"
- mutable
- unique keys but allow duplicate values so overall no duplicates
- ordered
- non-indexed as values can be access using keys in the dictionary

```python
dct={} ## empty dictionary
dct['name']='ritik' ## adding key-value pair to the dictionary
dct['age']=21
dct['college']='LIET'
print(dct)
print(type(dct))
dct1={'name':'ritik','age':21,'college':'LIET'} ## another way to create dictionary
print(dct1)
dct2={'student':{'name':'ritik','age':21,'college':'LIET','subjects':['python','java','c++']}} 
## values can be int, string or in the form of list, tuple,even dictionary itself
print(dct2)
```

Dictionary Operations : 

```python
dict={'name':'ritik','age':21,'college':'LIET'}
print(len(dict)) ## return the number of key-value pairs in the dictionary
print(dct['name']) ## access value of the key using key name
print(dct.get('age')) ## access value of the key using get() method
print(dct.keys()) ## return  keys of the dictionary
print(dct.values()) ## return values of the dictionary
print(dct.items()) ## return items (key-value pairs) of the dictionary
```

** Question : What will happen when you insert two same key but with different values ?

   Answer: As we know Dictionary doesn't allow any duplicacy of keys , keys must be unique. In this case dictionary takes first key with values and overwrite the value of key when you insert the same key-value pair.

```python
dict={'name':'ritik','age':21,'college':'LIET'}
print(dict)
dict2={'name':'ritik','name':'ritik kumar','age':21,'college':'LIET'} 
## if we have duplicate keys in the dictionary then value of first key-value pair overwritten by other one.
print(dict2)
```

```python
dict={'name':'ritik','age':21,'college':'LIET'}
print(dict)
dict.update({'grad_year':2023}) 
## used to update or add new item in the dictionary dict.update({key:value})
print(dict)
dict2=dict.copy() ## used to create a copy of the dictionary
print(dict2)
dict.pop('grad_year') ## used to remove specified key-value pair from the dictionary dict.pop(key)
## in dictionary pop need argument of key which we want to remove
print(dict)
dict.popitem() ## remove last item (key:value) pair from the dictionary
print(dict)
dict.clear() ## used to remove all items from the dictionary
print(dict)
print(dict2)
```

10. set : collection of unique elements enclosed within {}
- unordered
- unindexed
- mutable
- no duplicacy 

```python
sets={1,'hi',2.5,True,1}
print(sets) ## The values True and 1 are considered the same value in sets, and are treated as duplicates
print(type(sets))
set2={1,'hi',2.5,False,1}
print(set2)
set3={1,'hi',2.5,0,False} ## The values False and 0 are considered the same value in sets, and are treated as duplicates
print(set3)
```

Sets operations : 

```python
sets={1,'hi',2.5,True,1}
print(sets)
print(len(sets))
 
sets.add(6) ## used to add element in the set
print(sets)
sets.remove(6)
print(sets)
sets.discard(2) ## used to remove element from the set sets.discard(element)
print(sets)
sets.pop() ## used to remove random element from the set
print(sets)
set2={2,5,3}
sets.update(set2) ## used to add multiple elements in the set sets.update(set)
print(sets)
## you can also add list, tuple or dictionary in the set using update() method
lst=[7,8,9]
sets.update(lst)
print(sets)
sets.clear() ## used to remove all elements from the set
print(sets)
```

** Question: Difference between remove(),discard() and pop()?
   Answer: 
   1. remove() gives error when specified item/element not present in the set where discard() doesn't raise any error
   2. remove() and discard() are just used to remove specified item but they never return any value whereas pop() used to remove random item from set as well as return that removed value

```python
sets={1,2,3,4}
sets.remove(5)
```

```python
sets={1,2,3,4}
sets.discard(5)
print(sets.remove(2))
print(sets.discard(3))
print(sets.pop())
```

set operations on two sets :

```python
set1={1,2,3}
print(set1)
set2=set1.copy() ## used to create a copy of the set
print(set2)
set2.update((4,5))
print(set2)

## operations on two sets
print(set1.union(set2)) ## return a new set containing all unique elements from both sets
print(set1 | set2) ## | represents union
print(set1.intersection(set2)) ## return set of common elements
print(set1 & set2) ## & represents intersection
print(set1.difference(set2)) ## return set of elements present in set1 but not  
print(set2 - set1) ## - represents difference
```

** Question: sets are unindexed and unordered as well as not a key:value pair then how can you   access elements of sets.

Answer : we can't access single element  of set, we just check if specified element present in set or not and access all elements presents in the set using for loop.


```python
sets={1,'hi',2.5,True,1}
print(1 in sets) ## return true if 1 is present in set otherwise false
## in is the operator which is used to check the presence of element in the set

for i in sets:
    print(i) ## used to iterate through the set elements
```

11. FrozenSet : immutable version of set

```python
x={1,2,3}
print(type(x))
y=frozenset(x)
print(type(y))
print(y)
y.add(6) ## you can't add element in frozenset because it's immutable
print(y)
```

NOTE : all operations performed on sets can also be performed on frozen sets
